## Exercise 1: Cleaning Column Names

To ensure consistency and ease of use, standardize the column names of the dataframe.  
 Start by taking a first look at the dataframe and identifying any column names that need to be modified.  
  Use appropriate naming conventions and make sure that column names are descriptive and informative.

*Hint*:
- *Column names should be in lower case*
- *White spaces in column names should be replaced by `_`*
- *`st` could be replaced for `state`*

```python
import pandas as pd
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df = pd.read_csv(url)
df.head(10)
```

```python
df.shape
```

- Cambio la columna ST por STATE

```python
if "ST" in df.columns:
    df = df.rename(columns={"ST": "STATE"})
df
```

```python
df.columns = (
df.columns
  .str.strip()              # elimina espacios al inicio y final
  .str.lower()              # convierte a minúsculas
  .str.replace(' ', '_')    # reemplaza espacios por _
  )
df.head()
```

## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis.  
Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:



- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*

```python
df['gender'] = df['gender'].str.strip().str.lower()

df['gender'] = df['gender'].replace({
    'm': 'M', 'male': 'M',
    'f': 'F', 'female': 'F', 'femal': 'F'
})

df.head()
```

```python
df['gender'].value_counts() # Fantastico , queda F,M solo
```

- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*

```python
df['state'].unique()
```

```python
state_map = {
    'AZ': 'Arizona',
    'Cali': 'California',
    'WA': 'Washington',
}

df['state'] = df['state'].replace(state_map)
df.head()
```

- *In education, "Bachelors" could be replaced by "Bachelor"*

Primero vemos que valores hay en education

```python
df['education'].unique()
```

```python
df['education'] = df['education'].replace({'Bachelors': 'Bachelor'})
df.head()
```



- *In Customer Lifetime Value, delete the `%` character*


```python
df['customer_lifetime_value'].dtypes
```

```python
df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '')
```

```python
df.head()
```

- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
df['vehicle_class'].value_counts()
```

```python
df['vehicle_class'] = df['vehicle_class'].replace({
    'Sports Car': 'Luxury',
    'Luxury SUV': 'Luxury',
    'Luxury Car': 'Luxury'
})
df.head()
```
